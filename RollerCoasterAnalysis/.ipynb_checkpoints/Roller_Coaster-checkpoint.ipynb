```python
# 1 
# Import necessary libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# load rankings data
wood_rollercoaster = pd.read_csv('Golden_Ticket_Award_Winners_Wood.csv')
print(wood_rollercoaster.head(10))
# load rankings data
steel_rollercoaster = pd.read_csv('Golden_Ticket_Award_Winners_Steel.csv')
print(steel_rollercoaster.head(10))

```

For demonstration purposes, I have included examples using selected datasets, roller coasters, and attribute names of my choosing.

```python
# 2
# Function to plot rankings over time for 1 roller coaster
def rank_over_time(name, dataframe, park_name):
    coaster_data = dataframe[(dataframe['Name'] == name) & (dataframe['Park'] == park_name)]
    plt.plot(coaster_data['Year of Rank'], coaster_data['Rank'])
    plt.gca().invert_yaxis()
    plt.yticks([1,2,3])
    plt.xlabel('Year')
    plt.ylabel(f'Ranking of {name}')
    plt.title(f'Rankings for {name}')
    plt.show()
    plt.clf()

# A plot rankings over time for El Toro 
rank_over_time('El Toro', wood_rollercoaster, 'Six Flags Great Adventure')
```

Interpretation:

The graph shows the ranking of "El Toro" at Six Flags Great Adventure over the years.
We can observe how the coaster's performance fluctuated, with specific years where its rank improved or declined.
The ranking provides insights into how "El Toro" has been perceived in the roller coaster industry over time. A consistent high ranking could suggest continued popularity, while a drop in rank could indicate a decrease in its appeal compared to other coasters.


Summary:

"El Toro" is one of the top-ranked wooden roller coasters, consistently holding a position near the top over the years.
This analysis highlights the coaster's sustained excellence in the rankings, which can be used to emphasize its continued popularity in marketing or operational decisions.

```python
# Function to plot rankings over time for 2 roller coasters
def rank_over_time2(name, name2, dataframe, park_name, park_name2):
        coaster_data = dataframe[(dataframe['Name'] == name) & (dataframe['Park'] == park_name)]
        coaster_data2 = dataframe[(dataframe['Name'] == name2) & (dataframe['Park'] == park_name2)]
        plt.plot(coaster_data['Year of Rank'], coaster_data['Rank'], color = 'green', label = name)
        plt.plot(coaster_data2['Year of Rank'], coaster_data2['Rank'], color = 'red', label = name2)
        plt.gca().invert_yaxis()
        plt.yticks([1,2,3])
        plt.legend()
        plt.xlabel('Year')
        plt.ylabel(f'Ranking of {name} and {name2}')
        plt.title(f'Rankings for {name} and {name2}')
        plt.show()
        plt.clf()
        
# A plot of rankings over time for El Toro and Boulder dash hurricanes rollercoasters 
rank_over_time2('El Toro', 'Boulder Dash', wood_rollercoaster, 'Six Flags Great Adventure', 'Lake Compounce')
```

Interpretation:

The plot compares the rankings of two wooden roller coasters, "El Toro" and "Boulder Dash," over time.
"El Toro" (green line) is from Six Flags Great Adventure, and "Boulder Dash" (red line) is from Lake Compounce. The rankings are shown for both coasters over the years.
We observe how the rankings of both coasters have changed relative to each other, with each roller coaster's rank fluctuating at different points in time.
The y-axis, which is inverted, indicates that lower values represent higher ranks (1st place at the top), and the x-axis shows the year of ranking.


Summary:

This graph provides a direct comparison of the performance of two highly ranked roller coasters in the industry.
The analysis highlights how "El Toro" and "Boulder Dash" have competed for top positions over the years, with "El Toro" maintaining a consistently high ranking, while "Boulder Dash" shows more variability in its ranking.
This can be useful for understanding the relative success of these rides at their respective parks and assessing which coaster consistently performs better in the rankings over time.

```python
# 4
# Function to plot top n rankings over time

def top_ranking(n, dataframe):
    ranked_data = dataframe[dataframe['Rank'] <= n]
    plt.figure(figsize=(12,10)) 
    for coaster in ranked_data['Name'].unique():
        coaster_data = ranked_data[ranked_data['Name'] == coaster]
        plt.plot(coaster_data['Year of Rank'], coaster_data['Rank'], label=coaster)

    plt.gca().invert_yaxis()
    plt.legend(loc = 'lower right')
    plt.xlabel('Year')
    plt.ylabel(f'Rankings')
    plt.title(f'Top 10 Ranking')
    plt.show()
    plt.clf()
    
# A plot of top n rankings over time
top_ranking(5, wood_rollercoaster)
```

Interpretation:

This graph displays the rankings of the top 5 roller coasters over time. Each line represents a different coaster, showing how their ranks have evolved in relation to each other throughout the years.
The x-axis represents the year of ranking, while the y-axis represents the rank, where lower values correspond to higher rankings (1st place at the top).
The plot allows us to see the performance trends of various roller coasters in the top 5 ranks, with each coaster's line indicating how its ranking has changed year by year.

Summary:

This analysis provides a broader view of the most consistent and top-performing roller coasters over time.
We can see which coasters consistently maintain a high rank and which have seen more fluctuation in their positions.
The plot can be useful to stakeholders for understanding the long-term trends in the roller coaster industry, showing which rides are most frequently in the top rankings, and highlighting potential trends for future investments or marketing strategies.

```python
# 5
# load roller coaster data
roller_coaster = pd.read_csv('roller_coasters.csv')
print(roller_coaster.head(10))
```

```python
# 6
# Function to plot histogram of column values

# change column name for title and xlabel:

title_map = {
        'material_type': 'Material Type',
        'seating_type': 'Seating Type',
        'speed': 'Speed',
        'height': 'Height',
        'length': 'Length',
        'num_inversions': 'Number of Inversions',
        'manufacturer': 'Manufacturer',
        'park': 'Park',
        'status': 'Status'           
    }
    
 
    

def rollerecoaster_hist(dataframe, column_name):
    readable_title = title_map.get(column_name, column_name.replace('_', ' ').title())
    column_data = dataframe[column_name].dropna()
    if column_name == 'height':
        plt.hist(column_data, bins=10, range = (0,140), edgecolor = 'white', color = 'tomato')
    else:
        plt.hist(column_data, bins=10, edgecolor = 'white', color = 'tomato')
    plt.xlabel(readable_title)
    plt.ylabel('Count')
    plt.title(f'Histogram of Roller Coaster {readable_title}')
    plt.show()
    plt.clf()
    
# Histogram of roller coaster speed
rollerecoaster_hist(roller_coaster, 'speed')
# Histogram of roller coaster length
rollerecoaster_hist(roller_coaster, 'length')
# Histogram of roller coaster number of inversions
rollerecoaster_hist(roller_coaster, 'num_inversions')
# Histogram of roller coaster height
rollerecoaster_hist(roller_coaster, 'height')
```

Interpretation:

Histogram of Roller Coaster Speed:

This histogram shows the distribution of roller coaster speeds in the dataset. The majority of coasters have speeds in the range of 40–70 mph, indicating that most coasters are designed with moderate speeds, which may cater to a wide range of rider preferences.

Histogram of Roller Coaster Length:

This histogram highlights the distribution of roller coaster lengths. Most coasters fall within the 500 to 1500-meter range, suggesting that typical roller coasters are built with moderate lengths, likely to balance ride experience with space and cost constraints at parks.

Histogram of Roller Coaster Number of Inversions:

The histogram shows that the majority of coasters have no inversions, with a small number featuring multiple inversions. This suggests that non-inverting coasters are more prevalent, possibly indicating a preference for less intense rides.

Histogram of Roller Coaster Height:

The histogram for height reveals that most roller coasters have a height under 100 meters, with fewer coasters reaching extreme heights. The concentration of coasters in the lower height range could be indicative of practical and design limitations, as higher coasters often require more significant investment and engineering challenges.


Summary:

These histograms provide valuable insights into the characteristics of roller coasters in the dataset. From the distribution of speeds, lengths, inversions, and heights, we observe several key trends:

Moderate Speed: Most coasters are built with moderate speeds, balancing thrill with comfort.

Moderate Length: Coasters are generally designed to be compact yet fulfilling, fitting the available space while providing an enjoyable ride.

Non-inverting Designs: A majority of coasters do not have inversions, which could appeal to riders who prefer smoother experiences.

Practical Heights: Most coasters are built under 100 meters tall, focusing on a safer, more practical ride experience.

These insights are valuable when considering the design choices, target audience, and potential market for new roller coasters. These trends can also help identify opportunities for innovation, such as creating more extreme experiences (higher speeds or more inversions) or focusing on smoother rides for a broader audience.

```python
# 7
# Function to plot inversions by coaster at park

def num_inversions(dataframe, park_name):
    park_coasters = dataframe[dataframe['park'] == park_name]
#     print(park_coasters)

    park_coasters = park_coasters.sort_values('num_inversions', ascending=False)
#     print(park_coasters)

    coaster_names = park_coasters['name']
    number_inversions = park_coasters['num_inversions']
#     print(coaster_names)
#     print(number_inversions)

    plt.figure(figsize=(15, 6)) 
    plt.bar(range(len(coaster_names)), number_inversions, color = 'tomato')
    plt.title(f'Number of inversions of rollercoasters at {park_name}', fontsize = 16)
    plt.xlabel('Rollercoasters', fontsize = 14)
  # had to fix ticks doing the plt.bar way instead of .plot (kind=bar)
    ax = plt.subplot()
    ax.set_xticks(range(len(coaster_names)))
    ax.set_xticklabels(coaster_names, rotation = 60, ha='right')

    plt.ylabel('Number of inversions', fontsize = 14)
    plt.show()
    plt.clf()

num_inversions(roller_coaster, 'Six Flags Great Adventure')
```

Interpretation:

The bar chart displays the number of inversions for each roller coaster at Six Flags Great Adventure. The coasters are sorted in descending order, meaning the coasters with the most inversions are shown first.
The x-axis represents individual roller coasters, while the y-axis shows the number of inversions each coaster features.
The chart provides a clear visual comparison of inversion counts, allowing us to see which coasters are more intense in terms of inversions.

Summary:

This analysis helps to understand the variation in roller coaster design at Six Flags Great Adventure, particularly in terms of inversions. Coasters with a higher number of inversions generally provide a more thrilling and intense experience.
For park management, this information is useful when considering how to market different rides based on their intensity. For example, coasters with many inversions may appeal to thrill-seekers, while those with fewer inversions could attract families or less adventurous visitors.
The visualisation also highlights the diversity in ride experiences at the park, showing a range of coaster designs, from mild to extreme, catering to different guest preferences.

```python
# 8
# Create a function to plot a pie chart of status.operating

def status_operating(dataframe):
    operating_coasters = dataframe[dataframe['status'] == 'status.operating'] 
    closed_coasters = dataframe[dataframe['status'] == 'status.closed.definitely'] 
    status_counts = [len(operating_coasters), len(closed_coasters)]

    plt.figure(figsize=(10, 6))
    plt.pie(status_counts, autopct='%0.1f%%', labels = ['Operating', 'Closed'], colors = ['orange', 'coral'], textprops = {'fontsize': 11})
    plt.title('Operating vs Closed Roller Coasters', fontsize=14)
    plt.axis('Equal')
    plt.tight_layout() 
    plt.show()
    plt.clf()


# Create pie chart of roller coasters
status_operating(roller_coaster)
```

Interpretation:

This pie chart shows the distribution of roller coasters that are either operating or closed across the dataset.
The chart displays the percentage of roller coasters that are currently operating versus those that are closed permanently. The "Operating" category represents coasters still in service, while "Closed" refers to those that are no longer in operation.
The proportions are visualised clearly, allowing for easy comparison between the two groups.


Summary:

The pie chart provides stakeholders with an overview of the operational status of roller coasters in the dataset, revealing how many are currently in operation compared to those that have been permanently closed.
This information is useful for park management and industry analysis, as it reflects the number of active rides, which can influence decisions around investments, ride maintenance, and future attractions.
The high percentage of operating coasters suggests that the majority of roller coasters in the dataset are still functioning, highlighting a healthy state of ride availability for most parks.

```python
# 9
# Function to plot scatter of any two columns (removing outliers)

def scatter_plot(dataframe, column1, column2):
    plt.figure(figsize=(10,6))
    if column1 == 'height' or column2 == 'height':
        dataframe = dataframe[dataframe['height'] < 140]
    if column1 == 'speed' or column2 == 'speed':
        dataframe = dataframe[dataframe['speed'] < 250]
    plt.figure(figsize=(10,6))
    plt.scatter(dataframe[column1], dataframe[column2], color = 'tomato')
    plt.title(f'Scatterplot of {column1.replace("_", " ").title()} vs {column2.replace("_", " ").title()}', fontsize=16)
    plt.xlabel(column1.capitalize(), fontsize = 14)
    plt.ylabel(column2.capitalize(), fontsize= 14)
    plt.show()
    plt.clf()
    
# Scatter plot of speed vs height
scatter_plot(roller_coaster, 'height', 'speed')
```

Interpretation:

This scatter plot illustrates the relationship between height and speed for roller coasters in the dataset, after removing outliers (i.e., coasters with a height greater than 140m or speed greater than 250 mph).
The x-axis represents the height of the roller coasters, while the y-axis represents their speed.
From the plot, we can observe any trends or patterns that may exist between height and speed, helping to identify whether taller coasters tend to have higher speeds or if there is no significant correlation.


Summary:

The scatter plot offers a visual representation of how roller coaster height correlates with speed, showing that taller coasters generally do not necessarily correlate with faster speeds.
The data indicates that there are roller coasters with both high speeds and heights, as well as those with moderate speed and height. This could suggest that factors other than height, such as the design of the coaster and the technology used, may play a more significant role in determining speed.
This analysis is valuable in understanding the factors influencing ride experience and could help inform decisions related to roller coaster design, marketing, and investment, as well as how to appeal to different segments of park visitors.

```python
# What roller coaster seating type is most common? And do different seating types result in higher/faster/longer roller coasters?


def analyze_seating_types(dataframe):
    
    # To filter out true NaN and string representations of 'NaN'
    filtered_dataframe = dataframe[~dataframe['seating_type'].str.lower().str.strip().isin(['na', 'nan'])]
    
    # Group by seating type and count occurrences
    seating_counts = filtered_dataframe['seating_type'].value_counts()
#     print(seating_counts)
    
    # Calculate average height, speed, and length for each seating type
    seating_stats = filtered_dataframe.groupby('seating_type').agg({
        'height': 'mean',
        'speed': 'mean',
        'length': 'mean'
    }).reset_index()
#     print(seating_stats)
    
    # Plot the most common seating types
    plt.figure(figsize=(10, 6))
    seating_counts.plot(kind='bar', color='skyblue', edgecolor='black')
    plt.title('Most Common Roller Coaster Seating Types')
    plt.xlabel('Seating Type')
    plt.ylabel('Seating Type Count')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
    
     # To plot average height, speed, and length for each seating type (subplots)
    fig, axes = plt.subplots(1, 3, figsize=(18, 6))
    seating_stats.plot(x='seating_type', y='height', kind='bar', ax=axes[0], color='lightgreen', edgecolor='black')
    seating_stats.plot(x='seating_type', y='speed', kind='bar', ax=axes[1], color='lightcoral', edgecolor='black')
    seating_stats.plot(x='seating_type', y='length', kind='bar', ax=axes[2], color='lightblue', edgecolor='black')
    
    axes[0].set_title('Average Height by Seating Type')
    axes[0].set_xlabel('Seating Type')
    axes[0].set_ylabel('Average Height (m)')
    
    axes[1].set_title('Average Speed by Seating Type')
    axes[1].set_xlabel('Seating Type')
    axes[1].set_ylabel('Average Speed (km/h)')
    
    axes[2].set_title('Average Length by Seating Type')
    axes[2].set_xlabel('Seating Type')
    axes[2].set_ylabel('Average Length (m)')
    
    for ax in axes:
        ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')

        plt.tight_layout() 
    plt.show()
    
analyze_seating_types(roller_coaster)
```

Interpretation:

Bar Chart of Most Common Roller Coaster Seating Types:

This bar chart shows the distribution of seating types across all roller coasters in the dataset.
We can see which seating types are most prevalent in roller coaster designs. For example, "Sit Down" and "Suspended" seating types may dominate, suggesting these are the preferred seating arrangements in the industry.

Average Height by Seating Type:

This bar chart displays the average height of roller coasters for each seating type.
It provides insight into how different seating arrangements correlate with coaster height. For example, we might find that "Sit Down" coasters tend to be taller on average than "Suspended" or "Inverted" coasters.

Average Speed by Seating Type:

This bar chart shows the average speed for each seating type. It highlights the relationship between seating type and coaster speed.
Higher-speed coasters may correlate with certain seating types, providing a clearer picture of how speed influences design choices.

Average Length by Seating Type:

This bar chart shows the average length of roller coasters for each seating type. It provides further details about how the design of the coaster is influenced by the seating arrangement, particularly in terms of length.


Summary:

The analysis of seating types provides valuable insights into how different seating arrangements impact various coaster characteristics, such as height, speed, and length.

Common Seating Types: The most common seating types offer insight into industry trends and preferences, which can be useful for future roller coaster designs.

Design Considerations: The differences in average height, speed, and length by seating type suggest that certain seating configurations might be better suited for specific ride experiences, such as faster speeds or taller coasters. This data can guide decisions around which seating types are optimal for certain ride features, balancing comfort, thrill, and practicality.

This analysis can inform decisions on design strategies, marketing, and customer targeting, depending on the preferred seating types in the market.
