```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Example DataFrame (replace with your actual data from SQL)
data = {
    'segment': [87, 30, 87, 30, 87, 30],
    'month': ['2017-01-01', '2017-01-01', '2017-02-01', '2017-02-01', '2017-03-01', '2017-03-01'],
    'churn_rate': [25.1, 7.6, 31.7, 7.33, 47.7, 11.7]
}

df = pd.DataFrame(data)

# Convert 'month' to datetime format
df['month'] = pd.to_datetime(df['month'])

# Convert 'segment' to string for proper hue handling
df['segment'] = df['segment'].astype(str)

# Convert 'month' to string for proper handling in the legend
df['month'] = df['month'].dt.strftime('%Y-%m-%d')  # Convert to string

# ---------------------------------------------------
# Visualising churn rate by segment
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='month', y='churn_rate', hue='segment', 
            hue_order=['30', '87'], palette='Blues')  
plt.title('Churn Rate by Segment Over Time')
plt.xlabel('Month')
plt.ylabel('Churn Rate (%)')
plt.xticks(rotation=45)  # Rotate the month labels for readability
plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
plt.savefig('churn_rate_by_segment.png', dpi=300)  # Save first graph
plt.clf()

# ---------------------------------------------------
# Visualising churn rate comparison across segments (Bar chart) 
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='segment', y='churn_rate', hue='month', 
            order=['30', '87'], palette='Purples')  
plt.title('Churn Rate Comparison Across Segments')
plt.xlabel('Segment')
plt.ylabel('Churn Rate (%)')
plt.tight_layout()
plt.show()
plt.savefig('churn_rate_comparison.png', dpi=300)  # Save second graph
plt.clf()


plt.savefig('churn_rate_by_segment.png', dpi=300)

# to put in github portfolio readme file: ![Churn Rate by Segment](churn_rate_by_segment.png)
# 'churn_rate': [25.1, 7.6, 31.7, 7.3, 47.7, 11.7]
# plt.savefig('churn_rate_by_segment.png', dpi=300)  # Save first graph
#plt.savefig('churn_rate_comparison.png', dpi=300)  # Save second graph
```
